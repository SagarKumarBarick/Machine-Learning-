````python
# Boston Housing Price Prediction using Linear Regression

## Submitted By

**Name:** Sagar Kumar Barick
**Course:** Data Analytics / Machine Learning Basics
**Assignment:** ML Basics & Linear Regression Assignment

---

# Table of Contents

1. Introduction
2. Part A – Theoretical Questions
3. Part B – Practical Implementation
4. Exploratory Data Analysis
5. Model Building
6. Model Evaluation
7. Regression Equation
8. Interpretation of Coefficients
9. Conclusion

---

# Introduction

The objective of this project is to predict housing prices using the Boston Housing Dataset. Linear Regression is applied to understand how different housing features influence house prices and to build a predictive model.

---

# Part A – Theoretical Questions

## Q1. What is Machine Learning?

Machine Learning is a branch of Artificial Intelligence that enables systems to learn from data and make predictions or decisions without being explicitly programmed.

### Types of Machine Learning

### 1. Supervised Learning

Uses labeled data for training.

Examples:

* House Price Prediction
* Email Spam Detection

### 2. Unsupervised Learning

Uses unlabeled data to discover patterns.

Examples:

* Customer Segmentation
* Market Basket Analysis

### 3. Reinforcement Learning

Learns through rewards and penalties.

Examples:

* Self-driving Cars
* Game Playing AI

---

## Q2. What is Supervised Learning?

Supervised Learning is a machine learning technique where the model learns from labeled training data.

### Regression vs Classification

| Regression                 | Classification         |
| -------------------------- | ---------------------- |
| Predicts continuous values | Predicts categories    |
| Example: House Price       | Example: Spam/Not Spam |
| Output: Numeric            | Output: Class Label    |

---

## Q3. What is Linear Regression?

Linear Regression is a statistical technique used to model the relationship between independent variables and a dependent variable.

### Equation

Y = β₀ + β₁X₁ + β₂X₂ + ... + βₙXₙ

Where:

* Y = Target Variable
* β₀ = Intercept
* β₁...βₙ = Coefficients
* X₁...Xₙ = Features

### Intercept

The expected value of Y when all features are zero.

### Coefficient

Represents the change in Y for a one-unit increase in the feature.

---

## Q4. Evaluation Metrics

### SSE (Sum of Squared Errors)

Measures total squared prediction errors.

### MSE (Mean Squared Error)

Average of squared prediction errors.

### RMSE (Root Mean Squared Error)

Square root of MSE.

### R² Score

Measures proportion of variance explained by the model.

### Adjusted R²

Modified R² considering number of predictors.

---

## Q5. Machine Learning Workflow

1. Data Collection
2. Data Cleaning
3. Exploratory Data Analysis
4. Feature Selection
5. Train-Test Split
6. Model Building
7. Model Evaluation
8. Model Deployment

---

# Part B – Practical Implementation

## Q6. Load Dataset and Explore Data

### Python Code

```python
import pandas as pd

df = pd.read_csv("BostonHousing.csv")

print(df.head())
print(df.info())
print(df.describe())
```

### Observation

The dataset contains housing-related variables such as crime rate, number of rooms, tax rate, and house price (MEDV).

---

## Q7. Exploratory Data Analysis

### Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['rm','lstat','ptratio','medv']])
plt.show()
```

### Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

### Correlation Findings

Strong Positive Correlation:

* RM ↔ MEDV

Strong Negative Correlation:

* LSTAT ↔ MEDV
* PTRATIO ↔ MEDV

---

## Q8. Feature and Target Selection

```python
X = df.drop('medv', axis=1)
y = df['medv']
```

Target Variable:

* MEDV

Independent Variables:

* All remaining columns

---

## Q9. Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)
```

80% Training Data

20% Testing Data

---

## Q10. Linear Regression Model

### Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

### Prediction

```python
y_pred = model.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

n = len(y_test)
p = X_test.shape[1]

adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
```

### Output Metrics

Report the values obtained after execution:

* MAE = _______
* MSE = _______
* RMSE = _______
* R² = _______
* Adjusted R² = _______

---

# Regression Equation

After training:

```python
print(model.intercept_)
print(model.coef_)
```

General Equation:

MEDV =
β₀ +
β₁(CRIM) +
β₂(ZN) +
β₃(INDUS) +
β₄(CHAS) +
β₅(NOX) +
β₆(RM) +
β₇(AGE) +
β₈(DIS) +
β₉(RAD) +
β₁₀(TAX) +
β₁₁(PTRATIO) +
β₁₂(B) +
β₁₃(LSTAT)

---

# Interpretation of Coefficients

### RM

Positive coefficient indicates that more rooms increase house prices.

### LSTAT

Negative coefficient indicates that higher lower-status population percentage decreases house prices.

### PTRATIO

Negative coefficient suggests higher student-teacher ratio reduces property value.

### CRIM

Higher crime rates generally reduce housing prices.

---

# Conclusion

A Linear Regression model was successfully built using the Boston Housing Dataset. Exploratory Data Analysis revealed significant relationships between housing characteristics and property values. The model achieved satisfactory prediction performance and identified important factors such as number of rooms, crime rate, and lower-status population percentage as major determinants of house prices.
````

## Q6. Load Dataset and Explore Data

````python
import pandas as pd

df = pd.read_csv("BostonHousing.csv")

print(df.head())
print(df.info())
print(df.describe())
````

## Q7. Exploratory Data Analysis

### Pairplot

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['rm','lstat','ptratio','medv']])
plt.show()
````

### Heatmap

````python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
````

## Q8. Feature and Target Selection

````python
X = df.drop('medv', axis=1)
y = df['medv']
````

## Q9. Train-Test Split

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)
````

## Q10. Linear Regression Model

### Training and Prediction

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
````

### Evaluation

````python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

n = len(y_test)
p = X_test.shape[1]

adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
````

### Output Metrics

````python
print(f"MAE = {mae:.2f}")
print(f"MSE = {mse:.2f}")
print(f"RMSE = {rmse:.2f}")
print(f"R² = {r2:.2f}")
print(f"Adjusted R² = {adjusted_r2:.2f}")
````

# Regression Equation

````python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
````

### Formatted Regression Equation

````python
feature_names = X.columns
equation = f"MEDV = {model.intercept_:.2f}"
for i, coef in enumerate(model.coef_):
    equation += f" + ({coef:.2f})*{feature_names[i]}"
print(equation)
````
