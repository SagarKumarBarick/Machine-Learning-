# Section 1: Theoretical Questions

## Q1. What is a Decision Tree?
A Decision Tree is a supervised machine learning algorithm used for classification and regression.
- Root Node: Starting point of the tree.
- Internal Nodes: Decision points based on feature values.
- Leaf Nodes: Final prediction/output.

Example: A bank decides whether a customer will subscribe to a deposit based on age, balance, and previous campaign results.

## Q2. Gini Impurity vs Entropy
- Gini measures impurity using probability of misclassification.
- Entropy measures disorder using information gain.
- Scikit-learn uses Gini by default because it is computationally faster and usually provides similar performance.

## Q3. Overfitting
Overfitting occurs when a model learns training data too well and performs poorly on unseen data.
Detection:
- High training accuracy
- Significantly lower testing accuracy

## Q4. Pruning
Pruning reduces tree complexity.
- Pre-pruning: Stop tree growth early (max_depth, min_samples_split).
- Post-pruning: Grow full tree then remove unnecessary branches.

## Q5. Feature Importance
Feature importance shows which variables contribute most to predictions.
Businesses use it to identify the most influential factors affecting customer decisions.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

# Q6 Data Understanding
df = pd.read_csv('bank.csv')

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

df.head()
```

**Interpretation:** Dataset contains both numerical and categorical variables. The target variable is `deposit`.

```python

# Q7 Data Cleaning
print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

**Interpretation:** No major data quality issues. Missing values and duplicate counts should be verified from the output.

```python

# Q8 Data Preprocessing
le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

df.head()
```

**Encoding Method Used:** Label Encoding. Decision Trees require numerical inputs for model training.

```python

# Q9 Feature Selection & Splitting
X = df.drop('deposit', axis=1)
y = df['deposit']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(X_train.shape, X_test.shape)
```

**Interpretation:** Train-test split evaluates model performance on unseen data and prevents misleading results.

```python

# Q10 Model Building
model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=5,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

**Interpretation:** `max_depth=5` controls tree complexity and helps reduce overfitting.

```python

# Q11 Model Evaluation
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", acc)
print("\nConfusion Matrix:\n", cm)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

**Interpretation:** Accuracy, Precision, Recall and F1-score should be analyzed from the output. Precision measures correctness of positive predictions while Recall measures ability to identify actual positives.

```python

# Q12 Overfitting Check
train_acc = accuracy_score(y_train, model.predict(X_train))
test_acc = accuracy_score(y_test, y_pred)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```

**Interpretation:** If training accuracy is much higher than testing accuracy, the model is overfitting.

```python

# Q13 Pruning Experiment
pruned_model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
    min_samples_split=20,
    random_state=42
)

pruned_model.fit(X_train, y_train)

pruned_pred = pruned_model.predict(X_test)

print("Original Accuracy:", accuracy_score(y_test, y_pred))
print("Pruned Accuracy:", accuracy_score(y_test, pruned_pred))
```

**Interpretation:** Compare accuracies and choose the model that balances performance and generalization.

```python

# Q14 Feature Importance
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

print(importance.head())

importance.head(5).plot(
    x='Feature',
    y='Importance',
    kind='bar'
)

plt.title('Top 5 Feature Importances')
plt.show()
```

**Interpretation:** Features with the highest importance values influence predictions the most.
